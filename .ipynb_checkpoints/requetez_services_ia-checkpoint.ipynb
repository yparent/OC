# Projet OpenClassrooms – Requêtez des services IA

Objectif :
Tester un modèle de segmentation vestimentaire via l’API Hugging Face.

```python
import os
import time
import statistics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
from dotenv import load_dotenv
from huggingface_hub import InferenceClient
```

```python
load_dotenv()
HF_TOKEN = os.getenv("HF_TOKEN")

assert HF_TOKEN is not None
```

```python
model_id = "sayeed99/segformer_b3_clothes"
client = InferenceClient(model=model_id, token=HF_TOKEN)
```

```python
def run_segmentation(image_path):
    with open(image_path, "rb") as f:
        img = f.read()
    return client.image_segmentation(img)
```

```python
res = run_segmentation("data/test_images/img1.jpg")
res
```
